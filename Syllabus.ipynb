## COURSE OVERVIEW   
This course is based on and adapted from the course taught by Luigi Petrucco in the previous academic year (https://github.com/vigji/python-cimec-2025).   
The course aims to provide students with the foundational programming skills needed to use Python confidently for scientific work, with particular emphasis on writing, understanding, debugging, and organizing code and data-analysis workflows relevant to neuroscience research.

**Prerequisites**:
No prior programming experience is required. Basic familiarity with using a computer, managing files and folders, and installing software is recommended. A general scientific background and an interest in quantitative or neuroscience applications will be helpful.

**Learning outcomes**:
By the end of the course, students will be able to:

- use Python and Jupyter as practical tools for scientific programming,
- manage files, environments, and version control with the shell, conda, and GitHub,
- write and debug programs using core Python syntax, data structures, and control flow,
- analyze and manipulate numerical data with NumPy, Pandas
- structure simple coding projects in a clear and reproducible way, with applications to neuroscience.

**Technical requirements before the first lecture**

Before the start of the course, students are expected to have:

- a personal laptop (Windows, macOS, or Linux) with permission to install software (or have the software installed beforehand);
- **Miniconda** installed for Python/package management ([official installation guide](https://docs.anaconda.com/miniconda/)); 
- **Jupyter Notebook** or **JupyterLab** installed and executable on their machine ([Project Jupyter installation page](https://jupyter.org/install));
- a **GitHub** personal account already created and accessible ([GitHub sign-up page](https://github.com/signup));
- **one GitHub workflow option** already installed and ready to use (whichever you feel more comfortable with):
  - **GitHub Desktop** for a graphical interface ([official installation page](https://docs.github.com/en/desktop/installing-and-authenticating-to-github-desktop/installing-github-desktop));
  - or **GitHub CLI** / command line workflow ([official installation page](https://cli.github.com/));
- a web browser and a working internet connection.

Before the first class, students should verify that they can:

1. open a terminal;
2. run `conda --version`;
3. launch Jupyter with `jupyter lab` or `jupyter notebook`;
4. create a notebook and execute a simple Python cell.
5. log into GitHub successfully;
6. open either GitHub Desktop or GitHub CLI successfully.

## Course organization

The course is structured into **three modules**, each consisting of approximately **10-11 lectures** of **3 hours** each.

Each lecture is organized as a combination of **frontal teaching** and **hands-on coding practice**. Conceptual and methodological aspects will be introduced during the lecture component and then developed through guided practical work.

Practical activities will take place partly during class and partly as independent work at home. This organization is intended to support both guided learning and individual consolidation of the material.

The course will also include a **final project**, for which dedicated class time for discussion will be reserved. These sessions will allow students to work on their projects, discuss their progress, and receive feedback on both technical and organizational aspects of their work.

The schedule of topics is provisional and may be adjusted during the course according to the progress of the class.

## Assessment
Student assessment will be based on continuous practical work and a final project.   
The overall evaluation is distributed as follows:

- **40%** practical sessions
- **60%** final project, divided as follows:
  - **20%** project design
  - **20%** prompt formulation
  - **20%** coding

Assignments and final project materials must be submitted via **GitHub**.   
The course will be graded on a **pass/fail** basis.

**Using AI Tools**:
The use of generative AI tools is permitted in this course **within the limits of the University of Trento’s policy and the specific instructions provided by the instructor**.

In line with the University’s guidance, AI tools must be used in an **ethical, responsible, transparent, and academically honest** manner. Students remain fully responsible for the work they submit.

In this course, AI tools may be used to support activities such as:
- brainstorming;
- code explanation and debugging;
- code development and refinement;
- documentation support.

Any substantial use of AI tools in assignments or in the final project must be **explicitly declared** by the student. AI tools may not be used to conceal authorship, bypass learning objectives, or submit work that the student cannot understand and justify.

Further restrictions or task-specific rules may be communicated for individual assignments or project components.
