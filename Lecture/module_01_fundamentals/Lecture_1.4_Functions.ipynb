## Functions

A function is a **re-usable** block of code that performs operations on a specified set of **variables/parameters**, and returns the **result**.

### Anatomy of a function

**A function definition has**:
  - a <span style="color:indianred">name</span> that describes it, that we use to call it in the code (followed by `()`)
  - <span style="color:indianred">arguments</span> that we pass between the round brackets
  -  <span style="color:indianred">`return`ed values</span> that we can assign to new variables
  - (optional but strongly recommended): a <span style="color:indianred">docstring</span> (a documentation string)

**General form**

```python
def my_new_function(some_argument1, some_argument2):
    do something
    return something
```

### Arguments of a function

A function can have multiple input values:

````python
# write a function that print the exponentiation of a number by the other:
def print_something():
    print("Something!")

def exponentiate(a, b):
    return a**b

    
c = 2
d = 3
exponentiate(d, c)

print_something()
````

Some of the function arguments can have default values:

````python
# A function with default values:
def print_args(a, b, c=3):
    print(f"a={a}, b={b}, c={c}")

print_args(2,4,6)
````

We can pass the function values:
  - by position (**positional arguments**)
  - by keyword  (**keyword arguments**)
  
Positional arguments should precede keyword arguments!

````python
def print_args(a, b, c=2):
    print(f"a={a}, b={b}, c={c}")


# try passing stuff by position
print("By position, variable identity is inferred based on the arguments order")
print_args(1, 2, 3)

print("By keyword, we can specify separately each variable")
# try passing stuff by keyword:
print_args(c=1, a=2, b=3)
````

**Note**: positional and keyword arguments can be specified regardless of whether they are required or default-valued (in the function above, we can pass any of `a`, `b`, `c` by position or by value)

### Arguments type

- Python <span style="color:indianred">does not automatically enforce </span>the type of values passed to a function; what matters is whether an object supports the required operation.
  
- In modern Python, `type` hints can be added to suggest intended argument types and improve readability and tooling support.

- Hints are mainly for readability, editors, and static type checkers... **they do not enforce a data type**

````python
def double(x):
    return x * 2

print(double(3))      # 6
print(double("hi"))   # hihi
````

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">
Both work because double only relies on x supporting * 2, not on x having a specific declared type.</div>

If we use a <span style="color:indianred">hint </span>:

````python
def double(x: int) -> int:
    return x * 2

print(double(3))      # 6
print(double("hi"))   # hihi
````

- `def` starts a function definition.
- `double` is the function name.
- `x` is the input parameter.
- `: int` after `x` is a type hint: it says x is expected to be an integer.
- `-> int` is another type hint: it says the function is expected to return an integer.
- `return x * 2` sends back the value of x multiplied by 2.

### Naming functions

Function names should follow the same style as variable names:

- use **lowercase letters**
- separate words with **underscores**
- prefer names that clearly describe **what the function does**
- when possible, use a **verb**

````python
def foo():  # bad
    pass

def mean():  # good
    pass

def calculate_mean():  # better
    pass
````

A good function name should help the reader understand the purpose of the function **before** reading its code.

**Functions are objects**

In Python, functions can be treated like other objects.

This means that you can:

- **assign** them to variables
- **store** them in lists or dictionaries
- **pass** them as arguments to other functions of functions! 

````python
def multiply_numbers(x, y):
    return x * y

def add_numbers(x, y):
    return x + y

function_dict = {"multiply": multiply_numbers, "add": add_numbers}
````

Here, the dictionary does not store the *results* of the functions.   
It stores the **functions themselves**, which can be called late

````python
print(function_dict["multiply"](3, 4))   # 12
print(function_dict["add"](3, 4))        # 7
````

### Function docstrings

Functions should include a **docstring**: a short text that explains what the function does.

A docstring usually describes:

- the purpose of the function
- its arguments
- what it returns

This makes the function easier to understand and easier to use later.

````python
def exponentiate(a, b=2):
    """Compute a power using argument a as base and b as exponent.

    Parameters:
        a : int or float
            Base of the exponentiation
        b : int or float, optional
            Exponent of the exponentiation (default is 2)

    Returns:
        int or float
            The computed power
    """
    return a ** b
````

We can access the docstring directly in a notebook or IDE.   
This shows the function documentation without needing to open its code.

````python
?exponentiate
````

## Some useful functions - `map`
    map(function, iterable_object)

`map` applies the same function to every element in an iterable

- **Input**: a function + a sequence
- **Output**: the transformed elements, one by one

````python
numbers = [1, 2, 3]
result = map(str, numbers)

print(list(result))   # ['1', '2', '3']
````

## Some useful functions - `filter`
    filter(function, iterable_object)

`filter` selects the elements of an iterable that satisfy a condition and returns them as an iterator

- **Input**: a test function + a sequence
- **Output**: only the elements that pass the test

````python
def greater_than_two(x):
    return x > 2
    
numbers = [1, 2, 3, 4, 5]
result = filter(greater_than_two, numbers)

print(list(result))   # [3, 4, 5]
````

## Some useful functions - `lambda`
    lambda arguments: expression

Sometimes we write a small function only to use it once:

- **normal function**: give it a name, use it again later
- **lambda function**: short <span style="color:indianred">anonymous </span> function for <span style="color:indianred">one-time use</span> 


````python
numbers = [1, 2, 3]

result = map(lambda x: x**2, numbers)

print(list(result))   # [1, 4, 9]
````

### The scope of function variables

The **scope** of a variable is the part of the code where that variable exists and can be accessed.

Variables created inside a function are **local variables**:   
they only exist while the function is running.

````python
def sum_vals(a, b):
    local_sum_variable = a + b
    return local_sum_variable


a = 1
b = 2
d = sum_vals(a, b)

a = 2
b = 3
d = sum_vals(a, b)

print(local_sum_variable)
````

This fails because local_sum_variable only exists inside the function.

**Key concepts**:

- variables defined inside a function are **not available outside**
- they “live” only during the execution of the function

### Functions and external variables

Functions can access variables defined outside the function, but in general this should be avoided.

````python
def sum_vals(a, b):
    print(a_var_not_passed)
    return a + b

a = 1
b = 2
a_var_not_passed = [1, 2]
d = sum_vals(a, b)
````

This works, but it is not good practice.

Why?

Because the function now depends on something that **was not passed explicitly**.
If you move the function to another script or notebook, it may stop working.

**Good practice**:

A function should usually be as **independent** as possible.

So, whenever possible:

- pass needed values as arguments
- avoid relying on variables defined elsewhere in the notebook or script

## Side effects

A **side effect** happens when a function changes something outside its own local scope.

In general, we should avoid **side effects**, unless we really want them and make them explicit.

Typical side effects include:

-  an input object
- changing a variable outside the function
- writing to a file
- printing as part of the function behavior

````python
a_list = [1, 2, 3]

def take_last(a_list):
    a_list_copy = a_list.copy()
    return a_list_copy.pop()

a = take_last(a_list)
print("Last element is:", a)
print(a_list)
````

Here the original list is unchanged, because we made a **copy** first.

This is often a good strategy when we want to avoid modifying the input.

## In-place changes

Sometimes a function modifies the input object directly.   
This is called an **in-place change**.

````python
def double_list_values_inplace(input_list):
    """Double values inside a list in place."""
    for index, value in enumerate(input_list):
        input_list[index] = value * 2

num_list = [1, 10, 23]
print(num_list)
double_list_values_inplace(num_list)
print(num_list)
````

This is different from returning a new object:

````python
def double_list_values(input_list):
    return [val * 2 for val in input_list]
````

**Key distinction**:

- **in-place function** → changes the original object
- **non in-place function** → returns a new object

If a function changes its input in place, this should be made very clear:

- in the function name
- in the docstring
- in how you explain it to the reader
