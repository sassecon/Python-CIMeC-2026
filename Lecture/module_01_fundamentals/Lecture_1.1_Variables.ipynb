#### Slicing

Python allows getting slices from the string by calling

`string[start:end:step]` or `string[start:end]`

- `start` - index where you want to start the slicing (included)
- `end` - index where you want to end the slicing (not included)
- `step` - step of the slice (for example, every second value); default is 1

**Remember** indexes in Python start with 0

````python
s = "Neuroscience Rocks!"
print(len(s))   # returns the lenght of a string

print(s[5:12]) # from 5 to 12
print(s[5])    # 5th character
print(s[:4])   # from 0 to 4th
print(s[::2])  # every second value
print(s[::-1]) # returns reversed string
````

#### Check occurrence

To check whether a string holds some other string using the `in` operator.

**Note** strings (and Python in general) are case sensitive!! so "A" and "a" are two different strings.

````python
a_string = "The cat is called GARFIELD"
cat_name = "GARFIELD"
cat_name not in a_string
````

````python
cat_name in a_string
````

````python
cat_name = "Felix"
cat_name in a_string
````

**BUT**

````python
cat_name = "Garfield"
cat_name in a_string
````

#### Other operations with `str`

````python
s = "Neuroscience Rocks!"

# basic string methods (does not modify the original string)
s.lower()                  # returns 'neuroscience rocks!'
s.upper()                  # returns 'NEUROSCIENCE ROCKS!'
s.startswith('brain')      # returns False
s.endswith('!')            # returns True
s.isdigit()                # contains only decimal digits: returns False
s.find('science')          # returns index of first occurrence, which is 5
s.find('Psychology')       # returns -1 since not found
s.replace('Neuro','Brain') # replaces all instances of 'Neuro' with 'Brain'
s.split()                  # separates words

print(s)
````

Note that some (or even most) of the methods (aka functions), don't change the original object. In our example we applied multiple methods on the string s, but in the end it was still the same. If we want to save the modification after the method application we need to assign it to a variable.

Strings can incorporate variables values in a space defined by curly brackets with the **.format()** method:

## `==` vs `is`

Python has two different kinds of comparison:

- `==` checks whether two objects have the **same value**
- `is` checks whether two variables refer to the **same object**


````python
a_list = [1, 2, 3]
another_list = [1, 2, 3]

yet_another_list = a_list

print(a_list == another_list)       # True
print(a_list is another_list)       # False
print(yet_another_list is a_list)   # True
````

Why `x == None` is not ideal:  

`None` is a **singleton**

A singleton means:

- the object **exists only once**
- every reference **points to that same object**
- identity checks with `is` make sense
