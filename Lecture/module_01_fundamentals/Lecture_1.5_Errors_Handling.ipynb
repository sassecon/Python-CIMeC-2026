## Lecture outline

## Type of errors

Some errors happen before the program runs, while others happen during execution.

- **`SyntaxError`** is detected before execution because Python cannot parse the code.
- **`Exceptions`** usually happen at runtime, when Python is executing valid code.


> - If Python **cannot read** the code structure, you get a `SyntaxError`.
> - If Python can run the code but something **goes wrong while running** it, you get a runtime `exception`.

**Syntax errors**
  
**Definition**: Problems in the structure of your code that break Python’s rules.

**When they occur**: Before the code runs, when Python is checking the code for proper syntax.

- Missing colon after a statement
- Incorrect indentation
- Mismatched brackets, parentheses, or quotes

```python
if x > 5
    print("Hello")
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">This code will not run because the colon is missing.</div>

**Exceptions**
  
**Definition**: Errors that happen while the program is running, even if the syntax is correct.

**When they occur**: During runtime, after Python has started executing the code.

- `ZeroDivisionError`
- `NameError`
- `TypeError`

```python
print(10 / 0)
```

```python
print(name)
```

```python
print("5" + 2)
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">These lines raise exceptions because the operations cannot be completed as written.</div>

## Tracebakcs

**Exceptions** produce a traceback

When an exception is raised, Python prints a **traceback**.   
A traceback is a report of the function calls made in your code at a specific point. 

It helps you understand: 
- **what** error happened 
- **where** it happened 
- **which** sequence of calls led to it

```python
def greet(someone):
    print("Hello, " + someon)

greet("Anna")
```

### Anatomy of a traceback
![traceback](python_traceback.png)

- <span style="color:blue">exception name</span>
- <span style="color:green">information on the reason for the exception</span>
- <span style="color:orange">the sequence of function calls from the most recent to the oldest</span>
- <span style="color:indianred">the actual code that was executed</span>

#### `AttributeError`

> An `AttributeError` happens when you try to access an attribute or method that an object does not have.

In practice, this usually means:
- the attribute name is wrong,
- the method does not belong to that object type,
- you are using the wrong kind of object.

```python
my_string = "hello"
my_string.append("!")
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">append() is a method for lists, not for strings.</div>

#### `ImportError`

> An `ImportError` happens when Python cannot import a name, module, or object that your code asks for.

In practice, this usually means:
- the module does not contain that name,
- the import statement is written incorrectly,
- the requested object is not available where Python is looking.

```python
from math import squareroot
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">The math module exists, but squareroot is not a valid name in it. The correct function name is sqrt.</div>

#### `IndexError`

> An `IndexError` happens when you try to access a position in a sequence, such as a list or tuple, but that position does not exist (out of range).


```python
a_list = ['a', 'b']
a_list[3]
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">a_list has only two elements, at indexes 0 and 1.
Index 3 does not exist</div>

#### `KeyError`

> An `KeyError` happens when you try to access a dictionary key that does not exist. It means Python could not find the requested key in the dictionary.


```python
ages = {"Anna": 25, "Marco": 30}
ages["Luca"]
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">The dictionary contains "Anna" and "Marco", but not "Luca". The requested key is missing</div>

#### `NameError`

> A `NameError` happens when your code refers to a name that has not been defined. The undefined name can be a variable, function, or class name.


```python
def greet(someone):
    print("Hello, " + someon)

greet("Anna")
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">The function parameter is someone, but the code uses <strong>someon</strong>, which does not exist</div>

#### `SyntaxError`

> - A `SyntaxError` happens when Python finds code that does not follow Python’s syntax rules.   
> - This means Python cannot parse the code before running it.   
> - These errors are usually detected by the parser


```python
def greet(name)

    print("Hello", name)
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">A colon : is missing after the function definition.</div>

#### `TypeError`

> A `TypeError` happens when an operation or function is applied to an object of an inappropriate type.


```python
print("Age: " + 25)
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">`Age:` is a string, but `25` is an integer.
Python cannot join them directly with `+` </div>

#### `ValueError`

> - A `ValueError` happens when an operation or function receives an argument of the correct type but with an inappropriate value.
> - The type is acceptable, but the actual value is not.


```python
int("hello")
```

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">int() can convert strings, but only if they represent valid integers.
"hello" is a string, but not a valid number</div>

## Handling exceptions

- Python exceptions can be handled with `try` and `except`.
- Python runs the code inside `try`.
- If an exception occurs, Python jumps to the matching `except` block.
- If the exception is not caught, it propagates and Python prints a traceback.

```python
try:
    number = int("hello")
except ValueError:
    print("That is not a valid integer.")
```

- Catching exceptions lets the program respond to errors **more gracefully** instead of stopping immediately.
- Catch **specific exceptions** whenever possible, rather than using a generic `except`.

## Debugging workflow

![traceback](python_traceback.png) 

Read the traceback from **bottom to top**.

- Start from the <span style="color:green">**last line**</span>:
     - identify the exception type 
     - read the error message 
- Then move <span style="color:orange">**upward**</span>:
    - find the relevant file
    - check the line number
- inspect the <span style="color:indianred">**code and values**</span> involved
- follow the sequence of function calls if needed
