### Lists - `list`

- A **ordered** sequence of values of any type.  
- A list is a collection which is **ordered** and **changeable**.   
- Allows duplicate members.

````python
age = [15, 25, 45, 16, 18, 20, 25]
print(type(age))

len(age) # returns the length of the list

age[:5]     # returns [15, 25, 45, 16, 18]
age[3:7]    # returns [16, 18, 20, 25]
age[::-1]   # returns [25, 20, 18, 16, 45, 25, 15]
age[-2]     # returns 20
print(age)  # object stayed unchanged
````

#### Basic operations with `list`

````python
age = [15, 25, 45, 16, 18, 20, 25]

len(age) # returns 7
min(age) # returns 15
max(age) # returns 45
sum(age) # returns 164

avg_age = sum(age) / len(age)
print('avg_age = ',round(avg_age,2))
````

**Lists can hold objects of different types:**

(`sum` would not work in this case)

````python
a_list = [1, "Hello", [1,2,False], True-0, 42<3.14]
print(a_list)
````

The content of a list can be modified:

````python
a_list = [1, True, "something", 3.14]
a_list[2] = "new_content"
a_list
````

#### Other operations with `list`
Some of the methods are unique for lists.  
**Some** methods <font color="indianred">change</font>
 the initial string

````python
names = ['Bob', 'Bill', 'Sarah', 'Max', 'Jill']

# append one element to the end
names.append('Jack') 
print(names)
````

````python
# append multiple elements to the end
names.extend(['Anna', 'Bill']) 
print(names)
````

````python
# insert the element at index 0 (shifts everything to the right)
names.insert(0, 'Louis') 
print(names)
````

````python
# searches for the first instance and removes it
names.remove('Jill') 
print(names)
````

````python
# removes the element at index 1 and returns it
names.pop(1) 
print(names)
````

Some methods <font color="indianred">do not change</font> the initial string

````python
# methods that don't modify initial list and return a new object
print(names.count('Bill')) # returns the number of instances
print(names.index('Max'))  # returns the index of the first instance
````

### Tuples - `tuple`
Tuple is a collection which is **ordered** and **unchangeable**.  
Allows duplicate members.

We defined lists using the square brackets (`[1,2,3]`), but for tuples we use parentheses (`(1,2,3)`).

````python
brain_lobes = ('frontal', 'parietal', 'temporal', 'occipital')

print(brain_lobes)
print(type(brain_lobes))
````

````python
brain_lobes[0] = 'anterior'
````

Tuples are more limited than lists in terms of methods, and that is by design.  
- A tuple is **immutable**, which means that after it is created, its contents cannot be changed. 
- You cannot add, remove, or replace elements inside a tuple.  
- For this reason, **tuples have fewer methods than lists**

Used when grouping values that should not be modified.

### Sets - `set`
A `set` is a collection which is unsorted and unindexed. No duplicate members.  
Sets are defined using curly brackets `{1,2,3}`

````python
languages = {'python', 'r', 'java'} # create a set directly
snakes = set(['cobra', 'viper', 'python']) # create a set from a list
````

#### Basic operations with `set`
You can 
- loop through the set items using a for loop,
- ask if a specified value is present in a set, by using the `in` keyword,
- apply basic sets commands (like union or intersection)

````python
"python" in languages
````

````python
languages & snakes # intersection, AND
````

````python
languages | snakes # union, OR
````

````python
languages - snakes # set difference
````

 **Note** that we didn't get 'python' twice for the union since sets consist only of **unique values**

### Dictionaries - `dict`

Dictionaries are: **unordered**, iterable, **mutable**

- can contain multiple data types,
- is ordered with key-value pairs,
- keys can be strings, numbers, or tuples,
- the corresponding values can be any Python object

```
dict_obj = {
    'key1': value1,
    'key2': value2,
    ...}
```

- Dictionaries can be access by key (thus, don't allow duplicated keys)
- just keys or values can be accessed by `.keys()` and `.values()` methods

````python
participant = {'name': 'Jon Doe', 
               'group': 'Control', 
               'age': 42}

print(participant['name'])
````

````python
# add new key-value pair to the dictionary
participant['ID'] = 'CJD'
print(participant)
````

````python
# remove a key-value pair
participant.pop('ID')
print(participant)
````

````python
participant.keys()
````

````python
participant.values()
````

````python
participant.items()
````

````python
same_dict = dict(item_a=1, item_b=2)  # alternative way of defining a dictionary

print(same_dict)
print(same_dict["item_a"])
````

## Concatenated indexing

We can nest containers one inside the other! 
- lists of lists,
- dictionary of lists,
- dictionary of dictionaries...

In these cases, we can index nested elements adding **square brakets one after the other**:

````python
dict_of_lists = {"list_a": [0,1,2],
                 "list_b": [3,4,5]}

dict_of_lists["list_b"][0]
````

## <font color="indianred">**CAVEAT**</font>

**Careful!**: Changing the content of lists and dictionaries can have funny results!  
**Try**
- define a new variable from a list
- change the content of the original
- print the new variable

````python
a_list = [0, 1]
a_dictionary = {"a": a_list}
print(a_dictionary)
````

````python
a_list[1] = 2 
print(a_dictionary)
````

**The copy has also change!**
This, unless we use the `.copy()` method

````python
a_list = [0, 1]
another_list = a_list.copy()

print('a_list:', a_list)
print('another_list:', another_list)
````

````python
a_list[1] = 2

print('a_list:', a_list)
print('another_list:', another_list)
````

## Summary

### Python collectors
|Container |Syntax |Characteristics| Esample|
|---|---|---|---|
|List|`[]`|Ordered, mutable collection|`nums = [1, 2, 3]`
|Tuple|`()`|Ordered, immutable collection|`point = (3, 5)`|
|Set|`{}`|Unordered collection of unique elements|`colors = {"red", "blue"}`|
|Dictionary|`{key: value}`|Collection of key-value pairs|`student = {"name": "Sara", "age": 22}`|


### Quick comparison
- **List** → change elements, keep order
- **Tuple** → fixed data, keep order
- **Set** → unique items, no duplicates
- **Dictionary** → lookup by key
