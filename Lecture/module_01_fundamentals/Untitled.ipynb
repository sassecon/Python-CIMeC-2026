```python
subject_ids = ["s01", "s02", "s03", "s04"]
memory_scores = [21, 18, 25, 19]

all_ids = []
all_scores = []

for sid, score in zip(subject_ids, memory_scores):
    all_ids.append(sid)
    all_scores.append(score)

subject_record = {
    "subject_id": all_ids,
    "memory_score": all_scores
    }
print(subject_record)  
```
