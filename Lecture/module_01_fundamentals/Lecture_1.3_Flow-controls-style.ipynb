## Automating the routine: **LOOPS**

operations on variables and the flow of the program are managed with control structures:
 - `if` / `elif` / `else`
 - `for`
 - `while`

### `if` statement

make the execution of some lines of code optional:

```python
if condition is True:
    do something
elif another condition is True:
    do something
else:
    do something else

```

````python
x = 100
y = 500

if x > y:
    print('X is greater than Y')
elif x == y:
    print('X equals Y')
else:
    print('X is smaller than Y')
````

We can:
- combine **multiple conditions** in a boolean_result = `a > 1 and b < 10`
- specify an **alternative execution** with `else`
- specify **multiple alternative** executions with `elif`

````python
a = 6
if a < 5:
    print("smaller than 5")
elif a == 5:
    print("equal 5")
elif a > 5 and a < 10:
    print("bigger but smaller than 10")
elif a > 5 and a < 15:
    print("bigger but smaller than 15")
else:
    print("greater than 5")
````

### `while` loops
Using the `while` loop we execute the set of statements as long as condition is **True**.

```python

while condition is True:
    do something
    
```

````python
x = 0

while x < 4:
    x += 1 # which is the equivalent of `x = x + 1`
    print(x)
````

### `for` loops

With a `for` loop we can **repeat many times** the same code lines, iterating over a sequence (like lists, tuples, dictionaries, sets, strings).

```python

# iterate over all values in the list
for item in my_list:     
    print(item)
    
```

````python
stdev = [2, 4, 1.5, 2, 4]     # list of standard deviations

variances = []                # initialize the empty list that will hold variances
for val in stdev:             # iterate over all values in the list
    variances.append(val**2)  # append the variances list with squared value

print(variances)
````

## list comprehension

A **compact way** to create a new list by applying an expression to each item in an **iterable**, optionally filtering items with a condition.

```python

[expression for item in iterable]

[expression for item in iterable if condition]

```

```python

[expression for item in iterable]

```
translate to

```python

create empty list

for each item in iterable:
    
    compute expression using item
    
    append result to list
    
```  

````python
numbers = [1,2,3,4]
[x * 2 for x in numbers]
````

```python

[expression for item in iterable if condition]

```
translate to

```python

create empty list

for each item in iterable:
    
    if condition is true:
        
        compute expression using item
        
        append result to list
    
```  

````python
numbers = [-1,-2,3,4]
[x * 2 for x in numbers if x > 0]
````

#### Examples

````python
stdev = [2, 4, 0, 1.5, 2, 4]
variances = [val**2 for val in stdev]
print(variances)
````

````python
variances = [val**2 for val in stdev if val != 0]
print(variances)
````

````python
variances = [val**2 if val != 0 else "wrong value" for val in stdev]
print(variances)
````

list comprehension is

- nicer
- more efficient

<font color="indianred">**BUT**</font>  
**readability counts** --> avoid complicate nesting

## dictionary comprehension

A **compact way** to create a new dictionary by generating key: value pairs from an **iterable**, optionally with a condition.

```python

{key: val for something in something_iterable}

{key_expression: value_expression for item in iterable if condition}

```

**NOTE**: in `dict` iterable can be:
- key
- value
- item (pair key:value)

```python

{key_expression: value_expression for item in iterable}

```
translate to

```python

create empty dictionary

for each item in iterable:
    
    compute key
    compute value
    
    store key: value in dictionary
    
```  

````python
squares = {x: x**2 for x in range(5)}
print(squares)
````

```python

{key_expression: value_expression for item in iterable if condition}

```
translate to

```python

create empty dictionary

for each item in iterable:
    
    if condition is true:
        
        compute key
        compute value
        
        store key: value in dictionary
    
```  

````python
even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_squares)
````

## Iterate over iterables

- `enumerate`: **index** and **item** from a iterable
- `zip`: **two or more** iterables in parallel


### `enumerate`

````python
names = ["Anna", "Luca", "Maria"]

for i, name in enumerate(names):
    print(i, name)
````

### `zip`

````python
names = ["Anna", "Luca", "Maria"]
scores = [28, 30, 27]

for name, score in zip(names, scores):
    print(name, score)
````

## Breaking the loop

- `break`
- `continue`

Compare the output of the following code:

````python
stdev = [2, 4, 0, 1.5, 2, 4]

variances = []
for val in stdev: # example 1
    if val == 0:
        continue  # exit the current step of the loop
    variances.append(val**2)
print(variances)
````

````python
variances = []
for val in stdev: # example 2
    if val == 0:
        break  # exit the entire loop
    variances.append(val**2)
print(variances)
````

- `break`: exit the entire loop
- `continue`: exit the current step of the loop
