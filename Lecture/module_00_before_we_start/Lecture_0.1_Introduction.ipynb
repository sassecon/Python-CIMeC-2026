Named after the Monty Python's Flying Circus British TV show (not the animal)

### General purpose language

- wide range of applications (science, web, desktop applications, games, business...)
- support object-oriented programming, but also functional programming (arcane words that will soon make sense)
- used literally anywhere: Google, NASA, Meta, Reddit, Netflix, JP Morgan, Pixar...

### High level language

- Python (like MATLAB or R) is an <span style="color:indianred">interpreted language</span>
- Opposed to e.g. C, which is a <span style="color:indianred">compiled language</span>

**Compiled language**: all the instructions of the program are translated in assembly code (the computer language) before usage
 - slow to compile, but fast to run
 - generally, less user-friendly and harder to debug
 - code has to be compiled on every different machine

**Interpreted language**: each line of code is sent to an <span style="color:indianred">interpreter</span> that send instructions to the processor (using compiled code) and gives the response back
 - much slower than compiled code
 - user friendly and easier to interact with
 - code works across platforms

### Massive number of packages

 - very large (and quite geek) community
 - literally tens of thousands of packages to do pretty much anything...

### An additional Python strength: APIs

Many, many tools and websites offer Python <span style="color:indianred">APIs</span> (Application Programming Interfaces). With APIs we can:
 - download data from servers
 - publish content online
 - interact with other software / databases in our lab
 - build programs that do things in the world (read frames from cameras/sensors, control hardware, etc.)

### Coding with LLMs


- The quality of code from LLMs depend on availability of training data, and there is a lot of Python around!
- Increasing number of tools for programming with LLMs (Copilot, Cursor...)

## Using the terminal

- The terminal is a basic but very useful tool for working with a local Python installation

    - for Windows users, the terminal to use will be <span style="color:indianred">Anaconda prompt</span>* ;
    - for mac users, the <span style="color:indianred">Terminal</span>

- the terminal needs to know where the program is located, this is handled by system configuration such as the `PATH`

- The `PATH` is an environment variable
- It contains the list of folders where the terminal looks for executables
- You can print it in the terminal to see where commands such as `python` are searched for

`echo $PATH | tr ':' '\n'`

If a command is not in the `PATH`, the terminal cannot find it by name alone
- Use `./program` for a file in the current folder
- Use `/full/path/to/program` for a file in another folder

## Components of a Python installation

A local Python installation includes several parts:

- the **Python interpreter**
- the **Python standard library**
- **a package manager**
- **third-party packages**

These are the main elements that allow us to run Python code and extend Python with additional tools.

### Python interpreter
- is the program that runs Python code
- is an executable installed on your OS
- when we run `python`, we are calling this program

### Standard library
Python comes with many modules already included   

Examples: `random`, `os`, `sys`, `math`

The standard library provides many common tools out of the box.

### Package manager 
A **package manager** is used to install    
It helps us download and manage external libraries   
The most common package managers are:
- `pip`
- `conda`

### Third-party packages
Many useful libraries are **not** part of standard Python   
These are called **third-party packages**   

Examples: `numpy`, `pandas`   

These packages must be installed before they can be used.

### Optional but very useful parts

Some components are not strictly necessary, but are very useful in practice:

- an Integrated Development Evironment (**IDE**) or code editor
- an **environment manager**
- a **notebook runner**
- custom libraries

These tools make coding, organizing, and managing projects much easier.

#### IDEs and notebook tools
- Python code can be written in any text editor   
- **but** we usually use tools designed for coding

Examples: `Spyder`, `VS Code`, `PyCharm`

- `Jupyter` is a third-party package that allows us to open notebooks in the browser

#### Environment manager
Different projects may need different Python versions or package versions
- An **environment manager** keeps these setups separate
- avoids conflicts between projects
- Conda can also help manage environments


#### Custom libraries
- We can also create our own reusable Python modules
- These can contain functions or code used across multiple projects
- Once organized properly, they can be imported like other libraries
