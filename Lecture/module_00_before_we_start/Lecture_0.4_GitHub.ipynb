## The problem Git solves

- “Which file **is** the latest version?”
- “What changed between analysis_v7_final_final **and** v8?”
- “Can a collaborator **reproduce** my results?”
- “Can we track changes to preregistration, protocols, code, stimuli, **and** analyses?”

## Key concepts

- **Repository (repo)**: a project folder with history
- **Commit**: a snapshot + message (“what changed and why”)
- **Branch**: a parallel line of work (e.g., main, feature-cleaning)
- **Remote**: the GitHub copy of your repo
- **Push / Pull**: upload / download changes
- **Pull Request (PR**): propose changes + review
- **Clone**:  download the project + history

## What goes in a repo (and what doesn’t)

<span style="color:green"> **Good in Git**</span>:
- scripts (`.py, .R`), notebooks (`.ipynb`), docs (`.md`), small config files
- lightweight stimuli definitions, experiment code
- small sample data / synthetic data


<span style="color:red"> **Avoid in Git (use OSF/Zenodo/Drive)**</span>:
- large raw data (EEG/fMRI), big binaries, huge outputs

<div style="background:#e6f2ff; padding:20px; border-radius:12px;">
    
Use: data/ ignored + README explaining where to download data.

</div>

## GitHub workflow
We want to contribute or work on a repo that already exists


**1) First time only: get the repository on your computer** 
(if you do not already have the repository folder locally)
   ```bash
   git clone https://github.com/USERNAME/REPOSITORY.git
   cd REPOSITORY   
   ```

**2. Every time you start working**
- move into the folder
- download latest version from GitHub

```bash
cd REPOSITORY
git pull
```

**3. Work on your files locally**
- Python scripts
- notebooks
- markdown files
- project folders

**4. Check what changed**
```bash
git status
```

**5. Add your changes**
```bash
git add .
git add filename.py
```

**6. Commit your changes**
```bash
git commit -m "Update final project README"
git commit -m "Add NumPy practice notebook"
git commit -m "Fix syntax error in exercise 5"
```

**7. Upload your changes to GitHub**
```bash
git push
```


**Full daily workflow**

```bash
cd REPOSITORY
git status
git pull
# edit your files
git status
git add .
git commit -m "Describe your changes"
git push
```

## Commit messages

<span style="color:red">Bad</span>: update, final, fix

<span style="color:green">Good</span>:
- “Add preprocessing step for ICA (EEG)”
- “Document exclusion criteria in README”
- “Refactor Stroop trial generator; no behaviour change”

## Environment file for reproducibility
- For **conda**, one common solution is an environment file such as `environment.yml`.

- That lets another person rebuild a similar software environment.

In your repo:
```bash
conda env export --no-builds > environment.yml
git add environment.yml
git commit -m "Add conda environment for reproducibility"
git push
```


## `ReadMe.md` and `.gitignore`

Two files matter a lot early on:

- `README.md`: explains what the repository is
- `.gitignore`: tells Git which files to ignore

These are essential for good project managment.

### .gitignore (don’t commit junk)
A `.gitignore` file tells Git which files and folders it should **not track**.

How it works:

- Git reads the patterns listed in `.gitignore`
- if a file matches one of those patterns, Git ignores it
- ignored files do not show up as new untracked files in `git status`

Example `.gitignore`:

    data/
    results/
    __pycache__/
    .ipynb_checkpoints/
    *.csv
    *.nii
    *.edf

> **Adjust to your project!**

>**Important**:`.gitignore` only affects untracked files. If a file is already being tracked by Git, adding it to .gitignore will not automatically stop Git from tracking it

## Jupyter notebooks on GitHub (best practice)

- GitHub renders .ipynb nicely
- But notebooks can create noisy diffs

Recommendations:
- keep notebooks for reports/figures
- keep core logic in scripts (.py) and import them
- restart & run all before committing
