## What is an Operating System (OS)?

An **operating system** is:
- Core software that manages hardware and system resources
- Interface between **users**, **applications**, and **hardware**
- Manages **processes**, **memory**, **storage/files**, and **devices**

**Examples:** Windows, macOS, Linux, Android, iOS

## What is Unix?

**Unix** is a family of operating systems (developed in the 1970s, Bell Labs) and design ideas that strongly influenced modern systems such as **Linux** and **macOS**.

It’s designed to be **modular**: small tools can be combined efficiently.
It’s famous for the **command line interface** (CLI): interaction by typing commands.

Even if you are not using Unix directly, many command-line ideas come from the Unix world:
- files organized in directories
- small commands that do one thing well
- piping and redirection
- scripting

## Kernel vs shell

Two different words are important here:

- **Kernel**: the low-level core of the operating system
- **Shell**: a program that lets the user send commands to the system

The shell is not the OS kernel.

This distinction matters because later, in Jupyter, the word **kernel** will be used again with a different meaning.

## What is a SHELL

A **shell** is a program that provides a **command-line interface (CLI)** between you and the operating system.

You type **commands** as text, and the shell **interprets** them.

Examples of shells:
- `bash`
- `zsh`
- `powershell`

A **terminal** is the window or application where you interact with a shell.

## Shell vs Python vs Jupyter

- **Shell**: launches programs, moves files, sets environments  
- **Python**: does computation / analysis / modeling  
- **Jupyter**: notebook interface that runs Python via a **kernel**  

For example, we use the **shell** to start Jupyter from the **right environment**
> They are related, but they are not the same thing.

## “Kernel” can mean two different things

**OS kernel** (system kernel)
- the low-level core of the operating system

Jupyter / **Python kernel** (notebook kernel)
- A **running program** that connects your notebook code to`python` and manages execution
- The Python one is usually **IPython** (`ipykernel`)

## minimum shell vocabulary

- **Working directory**: “where you are” -> `pwd`
- **List files**: see what’s here -> `ls`
- **Change folder**: move around -> `cd path/`
- **Paths**
  - **Absolute**: starts from the root (e.g., `/Users/name/...`)
  - **Relative**: starts from here (e.g., `./data`, `../results`)
- **Command structure**
  - `command [options] [arguments]`
  - Example: `ls -l data/`
