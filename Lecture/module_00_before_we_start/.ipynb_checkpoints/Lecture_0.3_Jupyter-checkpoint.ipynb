## What is Jupyter?

**Jupyter** is an ecosystem for interactive computing.

A Jupyter notebook (`.ipynb`) mixes:
- code
- outputs
- formatted text
- equations
- images

In practice:
- **JupyterLab** or **Notebook** is the interface
- a **notebook** is the document you edit
- a **kernel** executes the code

### Notebook vs Kernel

These are different things:

- **Notebook** = the file and interface you are editing
- **Kernel** = the running computational engine that executes the code

When you run a code cell, the notebook sends the code to the kernel, and the kernel returns the output.

Each kernel is tied to a specific environment/interpreter (that’s why package availability depends on the kernel).

**Practical implications**:

 - Restart kernel = reset your session (variables, imports, state).
 - Change kernel = switch which Python/environment runs the notebook.

**Check your kernel here:**
`Kernel → Change Kernel → Python (Python-CIMeC)` (or your environment name)

**Jupyter in a browser** does not necessarily mean "online"

If you launch Jupyter from your own machine with:

```bash
jupyter lab
```

the interface opens in a browser, but the notebook is usually still running **locally** on your computer.

## Conda environments

A **conda environment** is an isolated Python installation with its own packages.


This is useful because:
- different projects may need different package versions
- **package conflicts** are common
- environments make **debugging** easier
- environments improve **reproducibility**

I will use: `Python-CIMeC`, give it your own name, make it recognisable


1) create environment (only once)   
   `conda create -n TempPerception`
2) activate environment   
   `conda activate TempPerception`
3) install the package needed   
   `conda install -c conda-forge spyder`
4) intstall multiple packages   
   `conda install -c conda-forge numpy pyedflib neurokit2 mne`
5) list the environments   
   `conda env list`


## Jupyter + Conda setup

### 1. Activate environment
- Check if conda is available
    `conda --version`
 - See existing environments 
    `conda env list`
 - Activate the course environment 
    `conda activate Python-CIMeC`

After this, your terminal prompt usually changes and shows the environment name.

`(Python-CIMeC) saraassecondi@Host-006 ~ %`

### 2. Installing packages

- activate the environment first, then install.
- Prefer conda when available:
`conda install numpy pandas matplotlib`
- Use `pip` only if conda cannot find the package:
`pip install some-package`
- Only use `pip` after activating Python-CIMeC, otherwise you may install into the wrong Python.

### 3. Start Jupyter from that environment


This ensures that Jupyter uses the correct:
- Python interpreter
- installed packages
- kernel

```bash
conda activate Python-CIMeC
jupyter-lab

### 4. Make the environment appear as a Jupyter kernel
- Sometimes the environment exists, but it does not appear in:   
`Kernel → Change Kernel`

- Run this once (inside the environment):      
`conda activate Python-CIMeC`    
`python -m ipykernel install --user --name Python-CIMeC --display-name "Python (Python-CIMeC)"`

- Then restart JupyterLab. You should now see the kernel:
`Python (Python-CIMeC)`

## Quick troubleshooting checklist

If a notebook behaves strangely, check:

- which kernel is selected?
- is the expected environment active?
- was the kernel restarted?
- was the notebook run out of order (cells executed in random order)?
- is the package installed in the same environment as the kernel?

## Mini cheat sheet 

|CONDA|descirption |
|---|---|
|`conda env list`| List all available environments|
|`conda activate myenv`| Switch into an environment|
|`conda deactivate`| Return to the base environment|
|`conda remove --name myenv --all`| Delete an environment entirely|
|`conda list`| Show packages installed in the active env|
|` Ctrl + C`| Stop
|`jupyter-lab`| Launch notebook|
|`python --version`| python version|
