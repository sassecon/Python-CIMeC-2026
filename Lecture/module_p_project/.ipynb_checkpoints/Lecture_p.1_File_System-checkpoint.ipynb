## Working with the filesystem

In Python, `Path` from the `pathlib` module

```python
from pathlib import Path

### Path objects
- A `Path` object represents a location in the filesystem
- It is created from a string containing a path
- The exact string depends on your machine and operating system



````python

from pathlib import Path

a_path = Path("/Users/name/code")
````

On Windows, backslashes can be problematic in strings
A raw string is often used:

````python
a_windows_path = Path(r"C:\Users\name\code")
````

### Checking whether a path exists
A Path can point to an existing or non-existing location
Use .exists() to check whether the path is present on disk

````python
from pathlib import Path

a_path = Path("/Users/name/project/file.txt")
print(a_path.exists())
````

This is useful before trying to read, move, or create files

### Useful Path attributes
Path objects provide useful information about files and folders

````python
from pathlib import Path

a_path = Path("/Users/name/project/Lecture1.0_Numpy-intro.ipynb")

print(a_path.name)    # file name
print(a_path.stem)    # file name without extension
print(a_path.suffix)  # file extension
print(a_path.parent)  # parent folder
print(a_path.parts)   # all path components
````

### Files vs folders

````python
from pathlib import Path

file_path = Path("/Users/name/project/file.txt")
folder_path = Path("/Users/name/project")

print(file_path.is_dir())
print(folder_path.is_dir())
````


### Browsing folders with glob()
- `.glob()` searches inside a folder for paths matching a pattern
- It is a convenient way to find files programmatically

````python

from pathlib import Path

folder = Path("/Users/name/project")
for path in folder.glob("*"):
    print(path)
````


`*` means “any string”

````python
Common glob() patterns
Find all Jupyter notebooks:
for path in folder.glob("*.ipynb"):
    print(path)
Find files whose name starts with Lecture1:
for path in folder.glob("Lecture1*"):
    print(path)
Find files whose name contains pandas:
for path in folder.glob("*pandas*"):
    print(path)
Patterns let us search by file name structure
````

````python
glob() and folders
glob("*") returns both files and folders
We can filter only directories using .is_dir()
for path in folder.glob("*"):
    if path.is_dir():
        print(path)
This is useful when we want to inspect only subfolders
````

````python
Recursive search with rglob()
.glob() searches only one folder level
.rglob() searches recursively inside subfolders as well
for path in folder.rglob("*.ipynb"):
    print(path)
This is useful for searching in large project trees
````

````python
glob() returns a generator
The result of .glob() is not a list
It is a generator: we can loop over it, but not index it directly
matches = folder.glob("*")
If we need indexing, convert it to a list:
matches = list(folder.glob("*"))
print(matches[0])
````

````python
Sorting paths
The order returned by glob() is not guaranteed
If order matters, sort the results explicitly
for path in sorted(folder.glob("*")):
    print(path)
This is important when processing files in a reproducible way
````

````python
Concatenating paths
pathlib lets us combine path parts with the / operator
This is portable across operating systems
from pathlib import Path

course_path = Path("/Users/name/code/python-course")
lectures_path = course_path / "lectures"
print(lectures_path
     This is safer than manually building path strings
````

````python

Creating folders
Use .mkdir() to create directories
from pathlib import Path

new_path = Path("/Users/name/new_folder")
new_path.mkdir(exist_ok=True)
exist_ok=True avoids an error if the folder already exists
````

````python
Slide 14 — Creating nested folders
Use parents=True to create missing parent directories as needed
from pathlib import Path

nested = Path("/Users/name/project/data/raw/session1")
nested.mkdir(parents=True, exist_ok=True)
This is useful for creating folder structures in one step
````

````python
Slide 15 — Moving files with .replace()
We can move a file or folder to a new location using .replace()
from pathlib import Path

destination_folder = Path("/Users/name/archive")
destination_folder.mkdir(exist_ok=True)

file_to_move = Path("/Users/name/report.txt")
file_to_move.replace(destination_folder / file_to_move.name)
This moves the file into the destination folder
````

````python
Automating file organization
If you repeatedly rename or move files manually, it is often better to automate it
Python can reorganize folders based on file names or metadata
from pathlib import Path

data_folder = Path("/Users/name/sample_data")
new_data_folder = Path("/Users/name/reorganized_data")
new_data_folder.mkdir(exist_ok=True)

for file in data_folder.glob("sub*"):
    subject, session, _, _ = file.stem.split("_")
    new_location = new_data_folder / subject / session
    new_location.mkdir(exist_ok=True, parents=True)
    file.replace(new_location / file.name)
````
