# Block Design in Neuroscience Coding Projects

How to design an analysis workflow before writing code

Start with a pipeline of blocks—not one giant script.

Question

Structure

QC

Preprocess

Output

Goal: learn to translate a scientific question into a clean, testable, reusable analysis workflow.


# WHY

Why design comes first

Without explicit design

With block design

- one giant notebook or script
- mixed preprocessing and statistics
- hidden assumptions about files or trials
- hard to debug or rerun later
- weak reproducibility

- clear data flow between stages
- reusable preprocessing and QC
- parameters kept explicit
- easier testing and debugging
- cleaner collaboration and replication

Design is the bridge between the scientific question and the implementation.


# CORE IDEA

What is a block?

A block is one functional unit with one main job, defined inputs, defined outputs, and a clear place in the pipeline.

INPUT

PROCESS

OUTPUT

files, arrays,
metadata, parameters

transform, clean,
compute, validate

cleaned data, metrics,
plots, saved artifacts

Every block should answer:
- What does it do?
- What does it need?
- What does it produce?

Rule of thumb: if you cannot describe a block in one sentence, it is probably too vague or too big.


# PRINCIPLES

What good block design looks like

One responsibility

Each block should do one main thing well.

Explicit interfaces

Inputs, outputs, and parameters are visible.

Independently testable

Problems should become visible at one block.

Reusable and swappable

A method can change without rewriting everything else.

Good design supports debugging, replication, collaboration, and later extension.


# WORKFLOW

A generic neuroscience project pipeline

Question

hypothesis, variable, comparison

Structure

subjects, sessions, metadata

Load

raw data and logs

QC

missing files, dimensions, alignment

Preprocess

filter, clean, normalize, segment

Features

ERP, firing rate, ROI, spectra

Stats

test, model, decode, compare

Output

figures, tables, serialized results

Key idea: each stage receives well-defined inputs and returns outputs that the next stage can trust.

- Which stages consume raw vs derived data?
- Which stages are reusable across studies?
- Where should checkpoints be saved or cached?


# LAYERS

From scientific question to code architecture

Model layer

Separate scientific intent from implementation details.


# TEMPLATE

Anatomy of a good block

Minimum specification

- name and purpose
- inputs and outputs
- parameters and assumptions
- validation tests
- saved artifact(s)

Example: EEG preprocessing block

preprocess_eeg(raw, cfg) → epochs

- reads raw recordings and parameters
- filters, re-references, epochs, baseline-corrects
- returns cleaned epochs for downstream analysis

A strong block is understandable without reading your mind.


# EXAMPLE

Worked example: EEG oddball / P300 study

Scientific question: Do target stimuli evoke a larger P300 than non-target stimuli?

1. Define inputs

subject IDs, files, events

2. Load EEG

recordings + trigger table

3. Check data

missing channels, event counts

4. Preprocess

filter, epoch, baseline

5. Extract ERP

target / non-target averages

6. Compare

P300 window statistics

- Likely outputs: cleaned epochs, subject-level amplitude table, group statistics, ERP plots.

# PRACTICE

Common mistakes and better design moves

Mistake

Better design move

- Start coding without a pipeline sketch

- Draw blocks and arrows first

- Put everything in one notebook or script

- Split responsibilities into functions or scripts

- Mix preprocessing and statistics

- Treat preprocessing as its own reusable stage

- Hide parameters in the middle of code

- Centralize parameters in one config section

- Fail to save intermediate outputs

- Save cleaned data, tables, figures, and logs


# APPLY IT

In-class exercise

You have spike trains from multiple neurons under two behavioral conditions.

Design the project as blocks before writing code.

- Name the blocks
- Specify inputs and outputs
- Decide what should be saved
- Identify where QC should happen

Suggested structure

Load → QC → Bin / smooth → Features → Compare → Output
