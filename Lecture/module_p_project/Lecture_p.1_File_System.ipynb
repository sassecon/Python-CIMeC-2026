### Path objects
- A `Path` object represents a location in the filesystem
- It is created from a string containing a path
- The exact string depends on your machine and operating system



````python

from pathlib import Path

a_path = Path("/Users/name/code")
````

On Windows, backslashes can be problematic in strings --> A raw string is used:

````python
a_windows_path = Path(r"C:\Users\name\code")
````


### Working (current) directory
In Python, that is represented by:

````python
here = Path(".")
# here = Path.cwd()

print(here)
print(here.resolve())
````

What this does:

- `Path(".")` means “the current directory”
- `resolve()` shows the full absolute path of that directory

So:
- `.` = here, current directory
- `..` = parent directory, one level up


### Browsing folders with glob()
- `.glob()` searches inside a folder for paths matching a pattern
- It is a convenient way to find files programmatically

````python

# main course folder
folder = Path("/Users/saraassecondi/Dropbox/workplace/University/Documents/Teaching/UNITN-Python_CIMeC/academic year 2025-2026/Python-CIMeC 2026/Lecture")
for path in folder.glob("*"):
    # `*` means “any string”
    print(path)
````

#### Common glob() patterns
Find all Jupyter notebooks:

````python
folder = Path("/Users/saraassecondi/Dropbox/workplace/University/Documents/Teaching/UNITN-Python_CIMeC/academic year 2025-2026/Python-CIMeC 2026/Lecture/module_00_before_we_start")

for path in folder.glob("*.ipynb"):
    print(path)
    
````

Find files whose name starts with Lecture_0.:

````python
for path in folder.glob("Lecture_0*"):
    print(path)
    
````

Find files whose name contains pandas:

````python
for path in folder.glob("*pandas*"):
    print(path)
````


#### glob() and folders
`glob("*")` returns both files and folders   
We can filter only directories using `.is_dir()`

````python
for path in folder.glob("*"):
    if path.is_dir():
        print(path)
````

#### Recursive search with rglob()
- `.glob()` searches only one folder level
- `.rglob()` searches recursively inside subfolders as well

````python
for path in folder.rglob("*.ipynb"):
    print(path)
````


#### glob() returns a generator
The result of `.glob()` is not a list
It is a generator: we can loop over it, but not index it directly

````python
matches = folder.glob("*")
````

If we need indexing, convert it to a list:

````python

matches = list(folder.glob("*"))
print(matches[0])
````

#### Sorting paths
The order returned by glob() is not guaranteed   
If order matters, sort the results explicitly (alphabetically)

````python
for path in sorted(folder.glob("*")):
    print(path)
````

### Creating folders
Use `.mkdir()` to create directories

````python
from pathlib import Path

new_path = course_path / "dummy"
new_path.mkdir(exist_ok=True)
````

`exist_ok=True` avoids an error if the folder already exists

````python
for path in course_path.rglob("du*"):
    print(path)
````

### Creating nested folders
Use `parents=True` to create missing parent directories as needed

````python
from pathlib import Path

nested = course_path / "dummy" / "dummySession1"
nested.mkdir(parents=True, exist_ok=True)
````

````python
folder = course_path / "dummy" 
for path in folder.rglob("*"):
    print(path)
````
